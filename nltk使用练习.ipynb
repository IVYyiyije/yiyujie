Tokenizing text into sentences

```python
para = "Hello World! It's good to see you!"
from nltk.tokenize import sent_tokenize
sent_tokenize(para) 
```

Tokenizing sentences into words

```python
from nltk.tokenize import word_tokenize
word_tokenize("hello world!")
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize('Hello World.')
```

```python
import nltk
text = "Hello World. Isn't this fun?"
pattern = r"\w+|[^\w\s]+"
print(nltk.tokenize.regexp_tokenize(text,pattern))
```

Tokenizing sentences using regular expressions

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer("[\w']+")
tokenizer.tokenize("Can't is a contaction.")
```

Filtering stopwords in a tokenized sentence

```python
>>> from nltk.corpus import stopwords

>>> english_stops = set(stopwords.words('english'))

>>> words = ["Can't", 'is', 'a', 'contraction']

>>> [word for word in words if word not in english_stops]
```

Calculating WordNet synset similarity

```python
>>> from nltk.corpus import wordnet as wn

>>> right = wn.synset('right_whale.n.01')

>>> minke = wn.synset('minke_whale.n.01')

>>> right.path_similarity(minke)
```

Stemming words

```python
from nltk.stem import PorterStemmer # "Poter"是一种词干提取的算法。

stemmer = PorterStemmer()

stemmer.stem('cooking')

```

```python
stemmer.stem('cookery')
```

```python
import nltk
```

```python
stemmer= nltk.PorterStemmer()
verbs = ['appears','appear','appeared','calling','called']
stems=[]
for verb in verbs:
    stemmed_verb=stemmer.stem(verb)
    stems.append(stemmed_verb)
```

```python
sorted(set(stems))
```
